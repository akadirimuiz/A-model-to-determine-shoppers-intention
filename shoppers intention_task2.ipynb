```python
shoppers_intention.replace({'False':0,'True':1},inplace=True)
```

```python
shoppers_intention['Revenue'].head()
```

```python
shoppers_intention.columns
```

```python
#converting categorical columns to numerical
categorical =['VisitorType','Month']
```

```python
#one hot enlcoding the categorical data
categorical_dummies= pd.get_dummies(shoppers_intention[categorical])
categorical_dummies.head()
```

```python
#concatenate the categorical with dataframe
shoppers_intention = pd.concat([categorical_dummies,shoppers_intention],axis=1)
print(shoppers_intention.shape)
shoppers_intention.head()

```

```python
#deleting the categorica variables after it has been one encoded
shoppers_intention = shoppers_intention.drop(categorical,axis=1)
shoppers_intention
```

```python
shoppers_intention.drop(['Administrative','Region','ProductRelated','BounceRates','PageValues','OperatingSystems','Browser','TrafficType'],axis=1)
shoppers_intention
```

```python
shoppers_intention.columns
```

```python
#selecting the target column
y=shoppers_intention.Revenue
x=shoppers_intention.drop('Revenue',axis=1)
```

```python
print(x)
print(y)
```

```python
#import the necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
#splitting the dataset
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.4)
```

```python
print('\n', x_train.head())
print('\n', x_test.head())
print('\n', y_train.head())
print('\n', y_test.head())
```

```python
#initialize the model
DT_model = DecisionTreeClassifier()
```

```python
#train the model 
DT_model = DT_model.fit(x_train,y_train)
```

```python
DT_preads = DT_model.predict(x_test)
```

```python
print(classification_report(DT_preads,y_test))
```

```python
#TRAIN AND TEST WITH ANOTHER MODEL:RANDOM_FOREST
RF_model = RandomForestClassifier()
```

```python
#train the model
RF_model = RF_model.fit(x_train,y_train)
```

```python
#make prediction
RF_pred = RF_model.predict(x_test)
```

```python
#determine the accuracy,precision and recall
print(classification_report(RF_pred,y_test))
```

Summary: In machine learning the accuracy_score tells how many times the machine learning model is correct overall. precision tells how good the machine learning model at predicting specific category while recall tells how many times the model can detect a specific categoory. this exercise focuses on specific category which is customers "purchase-intention" which translates to revenue generation as the core aim of every business is to make profit. The model with the highest precision will be the preferred model to adopt. The decision tree model has a higher precision score this will the preferred model.
